# 00 — My Working Directory

Before we do anything spatial, we must know:

- What is the current working directory?
- Does the `data/` folder exist?
- If the data doesn't exist, then stop nicely (don't crash). 
- We can fix it this time by creating and copying whats needed to your local folder.


```python
from pathlib import Path

cwd = Path.cwd()
print("Current Working Directory:")
print(cwd)

data_path = cwd / "data"
print("\nLooking for data folder at:")
print(data_path)

exists = data_path.exists()
print("\nDoes data folder exist?")
print(exists)

assert exists, (
    f"\n❌ ERROR: 'data/' folder not found at:\n{data_path}\n"
    "Check your working directory or update the path to find your file."
)

print("\n✅ Environment looks correct.")
```
